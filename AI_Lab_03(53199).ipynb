```python
import numpy as np

# 1. Create a 3x3 matrix with values ranging from 2 to 10
matrix = np.arange(2, 11).reshape(3, 3)
print("Original 3x3 Matrix:\n", matrix)

# 2. Convert the array to a float type
float_matrix = matrix.astype(float)
print("\nMatrix converted to Float:\n", float_matrix)

# 3. Flatten the array
flattened_array = float_matrix.reshape(-1)
print("\nFlattened Array:\n", flattened_array)

# 4. Find the maximum and minimum values
max_val = flattened_array.max()
min_val = flattened_array.min()

print(f"\nMaximum Value: {max_val}")
print(f"Minimum Value: {min_val}")
```
