```python
# ---------------------------------
# Depth First Search (DFS)
# ---------------------------------

def dfs(node, graph, visited):
    visited[node] = True          # Mark current node as visited
    print(node, end=" ")          # Print the node (DFS order)

    # Visit all adjacent (connected) nodes
    for neighbor in graph[node]:
        # If the neighbor is not visited, call DFS on it
        if not visited[neighbor]:
            dfs(neighbor, graph, visited)


# Main Program
if __name__ == "__main__":

    # Adjacency List Representation of Graph
    graph = {
        0: [1, 2, 3],
        1: [0],
        2: [0, 3, 4],
        3: [0, 2],
        4: [2]
    }

    # Print adjacency list
    print("Adjacency List:")
    for node in graph:
        print(f"{node} -> {graph[node]}")

    # Create visited list (Initially all False)
    visited = [False] * len(graph)

    print("\nDFS Traversal:")
    dfs(0, graph, visited)   # Start DFS from node 0
```

```python
# ---------------------------------
# Breadth First Search (BFS)
# ---------------------------------

from collections import deque   # Queue is required for BFS

def bfs(start, graph):

    visited = [False] * len(graph)   # Initially all nodes are unvisited
    queue = deque()                  # Create empty queue

    visited[start] = True            # Mark start node visited
    queue.append(start)              # Add start node to queue

    # Repeat until queue becomes empty
    while queue:
        node = queue.popleft()       # Remove front element
        print(node, end=" ")         # Print node (BFS order)

        # Visit all adjacent nodes
        for neighbor in graph[node]:
            # If not visited, mark visited and add to queue
            if not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)


# Main Program
if __name__ == "__main__":

    # Adjacency List Representation of Graph
    graph = {
        0: [1, 2, 3],
        1: [0],
        2: [0, 3, 4],
        3: [0, 2],
        4: [2]
    }

    # Print adjacency list
    print("Adjacency List:")
    for node in graph:
        print(f"{node} -> {graph[node]}")

    print("\nBFS Traversal:")
    bfs(0, graph)   # Start BFS from node 0
```
