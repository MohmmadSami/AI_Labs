```python
#Tasks:
#Update puzzle values, size, analyze the code and upload on Moellim.

import heapq   # Used to implement priority queue

# Goal state we want to reach
goal_state = [[1,2,3],
              [4,5,6],
              [7,8,0]]   # 0 represents the empty space

# Possible moves of empty tile (up, down, left, right)
moves = [(-1,0),(1,0),(0,-1),(0,1)]


# Heuristic Function (Manhattan Distance)
# Calculates estimated distance from current state to goal state
def heuristic(state):
    distance = 0

    # Traverse the puzzle grid
    for i in range(3):
        for j in range(3):
            value = state[i][j]

            # Ignore the empty tile
            if value != 0:
                # Find the correct position of the tile in goal state
                goal_x = (value-1)//3
                goal_y = (value-1)%3

                # Manhattan Distance formula
                distance += abs(i-goal_x) + abs(j-goal_y)

    return distance


# Function to find the position of the empty tile (0)
def find_zero(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i,j


# A* Search Algorithm
def astar(start):

    pq = []   # Priority queue
    heapq.heappush(pq,(0,start))   # Push initial state with cost 0

    visited = set()   # To store visited states

    while pq:

        # Pop the state with lowest f(n)
        f,state = heapq.heappop(pq)

        # Check if goal state is reached
        if state == goal_state:
            print("Goal Reached")
            for row in state:
                print(row)
            return

        # Mark state as visited
        visited.add(str(state))

        # Find location of empty tile
        x,y = find_zero(state)

        # Try all possible moves
        for dx,dy in moves:

            nx,ny = x+dx , y+dy   # New position after move

            # Check boundaries of grid
            if 0<=nx<3 and 0<=ny<3:

                # Create a copy of current state
                new_state = [row[:] for row in state]

                # Swap empty tile with adjacent tile
                new_state[x][y],new_state[nx][ny] = new_state[nx][ny],new_state[x][y]

                # If state not already visited
                if str(new_state) not in visited:

                    g = 1  # Cost from start (for simplicity we take 1 step cost)

                    # Heuristic estimate to goal
                    h = heuristic(new_state)

                    # Total cost function
                    f = g + h

                    # Push new state into priority queue
                    heapq.heappush(pq,(f,new_state))


# Initial puzzle state
start_state = [[1,2,3],
               [4,0,6],
               [7,5,8]]

# Run A* algorithm
astar(start_state)
```

```python
#Task:
#	Implement A* example other than above example. i.e shortest path of graphs, maps, games.

import heapq   # Used for priority queue


# Graph representation using dictionary
# Each node contains neighbors with edge cost
graph = {
    'A': [('B',1),('C',3)],
    'B': [('D',3),('E',6)],
    'C': [('F',5)],
    'D': [],
    'E': [('F',1)],
    'F': []
}


# Heuristic values (estimated distance to goal)
heuristic = {
    'A':7,
    'B':6,
    'C':2,
    'D':1,
    'E':1,
    'F':0
}


# A* Search Function
def astar(start, goal):

    pq = []   # Priority queue

    # Push start node with initial cost
    heapq.heappush(pq,(0,start))

    cost = {start:0}   # Store actual cost from start node
    parent = {}        # To reconstruct the path


    while pq:

        # Get node with lowest f(n)
        f,current = heapq.heappop(pq)

        # If goal reached, reconstruct path
        if current == goal:

            path=[]

            while current in parent:
                path.append(current)
                current = parent[current]

            path.append(start)
            path.reverse()

            print("Shortest Path:",path)
            return


        # Explore neighbors
        for neighbor,weight in graph[current]:

            # Calculate new cost from start
            new_cost = cost[current] + weight

            # If node not visited OR better cost found
            if neighbor not in cost or new_cost < cost[neighbor]:

                cost[neighbor] = new_cost

                # f(n) = g(n) + h(n)
                f = new_cost + heuristic[neighbor]

                # Push into priority queue
                heapq.heappush(pq,(f,neighbor))

                # Save parent to track path
                parent[neighbor] = current


# Run algorithm from A to F
astar('A','F')
```
