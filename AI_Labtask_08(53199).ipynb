```python
import pandas as pd
import numpy as np
from io import StringIO

# ── Dataset ──────────────────────────────────────────────────────────────────
csv_data = """StudentID,Name,Gender,Age,Department,GPA,Attendance,Quiz1,Quiz2,FinalExam
1,Ayesha,Female,20,CS,3.7,85,8,9,88
2,Ali,Male,22,SE,2.9,72,5,6,70
3,Sara,Female,21,CS,3.2,95,7,8,83
4,Usman,Male,23,IT,2.5,65,6,5,68
5,Anum,Female,20,CS,3.9,90,9,10,92
6,Ahmed,Male,24,SE,2.7,60,4,5,63
7,Hira,Female,22,IT,3.4,80,7,8,79
8,Kamran,Male,23,CS,3.0,75,6,6,72
9,Mahnoor,Female,21,SE,3.5,88,8,9,87
10,Zain,Male,22,CS,2.8,55,5,4,60"""

df = pd.read_csv(StringIO(csv_data))
print('Dataset loaded successfully!')
df
```

## 📊 Task 1 — Data Exploration

```python
# Summary statistics (numeric columns only)
print('=== describe() ===')
display(df.describe())

# Full structure: dtypes, non-null counts, memory
print('\n=== info() ===')
df.info()
```

## 🧹 Task 2 — Data Cleaning

```python
# Remove duplicate rows (none expected, but good practice)
df.drop_duplicates(inplace=True)
print(f'Duplicates removed. Rows remaining: {len(df)}')

# Check for missing values
print('\nMissing values per column:')
print(df.isnull().sum())

# Fill any missing numeric values with column mean
df.fillna(df.select_dtypes(include='number').mean(), inplace=True)

# Rename FinalExam → Final_Exam_Score
df.rename(columns={'FinalExam': 'Final_Exam_Score'}, inplace=True)
print('\nColumn renamed. Current columns:', df.columns.tolist())
```

## 🎯 Task 3 — Feature Engineering

```python
# Performance_Level based on GPA thresholds
df['Performance_Level'] = pd.cut(
    df['GPA'],
    bins=[0, 2.99, 3.49, 5],
    labels=['Needs Improvement', 'Good', 'Excellent']
)

# Total quiz score
df['Total_Quiz'] = df['Quiz1'] + df['Quiz2']

print('New columns added:')
display(df[['Name', 'GPA', 'Performance_Level', 'Quiz1', 'Quiz2', 'Total_Quiz']])
```

## 📈 Task 4 — Sorting & Grouping

```python
# Sort by GPA descending, then Attendance ascending
sorted_df = df.sort_values(by=['GPA', 'Attendance'], ascending=[False, True])
print('=== Sorted DataFrame ===')
display(sorted_df[['Name', 'Department', 'GPA', 'Attendance']])

# Group by Department → avg GPA, avg Attendance, student count
grouped = df.groupby('Department').agg(
    Avg_GPA       = ('GPA', 'mean'),
    Avg_Attendance= ('Attendance', 'mean'),
    Student_Count = ('StudentID', 'count')
).round(2)

print('\n=== Department Summary ===')
display(grouped)
```

## 🧪 Bonus — Students Above Department Avg Final Exam Score

```python
# Calculate each department's average Final_Exam_Score
dept_avg = df.groupby('Department')['Final_Exam_Score'].transform('mean')

# Filter students who scored above their department's average
top_students = df[df['Final_Exam_Score'] > dept_avg][['Name', 'Department', 'Final_Exam_Score']]

print('Students above their department average Final Exam Score:')
display(top_students)
```
